1. Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load Dataset

```
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

3. Create Visualizations

a. Survival Count Plot

```
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

b. Gender-wise Survival

```
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Gender-wise Survival")
plt.show()
```

c. Passenger Class Distribution

```
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

d. Age Distribution

```
plt.hist(df['Age'].dropna(), bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

e. Fare Distribution

```
sns.histplot(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.show()
```

f. Correlation Heatmap

```
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

4. Conclusion

Through various visualizations, we identified important trends and relationships in the Titanic dataset. Survival rates were higher among females and first-class passengers. Visual analysis helped transform raw data into understandable insights and improved decision-making through graphical representation.
